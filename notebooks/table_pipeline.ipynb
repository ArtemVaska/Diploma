```python
import pandas as pd

from datasets import select_phyla, download_all_files_ncbi
```

```python
column_names = ["tax_id", "org_name", "gene_id", "current_id", "status", "symbol", "aliases", "description",
                "other_designations", "map_location", "chromosome", "genomic_nucleotide_accession.version",
                "start_position_on_the_genomic_accession", "end_position_on_the_genomic_accession", "orientation",
                "exon_count", "to_delete_1", "to_delete_2"]

df = pd.read_csv("../all_nxf1_2.txt", sep="\t", skiprows=1, names=column_names, index_col=0)
df.drop(["to_delete_1", "to_delete_2"], axis=1, inplace=True)
df.head()
```

```python
phylas = [
    "Porifera",
    "Cnidaria",
    "Ctenophora",
    "Platyhelminthes",
    "Nematoda",
    "Annelida",
    "Mollusca",
    "Arthropoda",
    "Echinodermata",
    "Bryozoa",
    "Rotifera",
    "Nemertea",
    "Tardigrada",
    "Onychophora",
    "Sipuncula",
    "Brachiopoda",
    "Priapulida",
    "Loricifera",
    "Kinorhyncha",
    "Gastrotricha",
    "Chaetognatha",
    "Choanoflagellata"
]
```

```python
# long-time func !!!
phyla_taxids, taxids_error = select_phyla(df, phylas)
phyla_taxids
```

```python
with open("../found_phyla_taxids.txt", "w") as outfile:
    for phyla, taxids in phyla_taxids.items():
        outfile.write(f"{phyla}:{taxids}\n")
```
